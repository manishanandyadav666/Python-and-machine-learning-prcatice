```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(0,10)
```

```python
#Plotting using matplotlib
plt.scatter(x,y,c='r')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph in 2D')
plt.savefig('test.jpg')
```

```python
y=x*x
plt.plot(x,y,'r|')
```

```python
#Creating subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*')
plt.subplot(2,2,3)
plt.plot(x,y,'b+')
plt.subplot(2,2,4)
plt.plot(x,y,'yo')
```

### Pass by value and pass by reference

### In python, neither pass by reference nor pass by value. Integer, str are immutable which is different form list

```python
## Same id at the begning and id changes after change in variable
def modify(x):
    print(id(x))
    x=5
    print("id of x", id(x))
    print("Value of x", x)
a=7
modify(a)

print("Vaue of a", a)
print("id of a", id(a))
```

```python
## Difference when the mutable type is used
def modfy(lst):
    print(id(lst))
    lst[2]=6
    print(id(lst))
    print('list', lst)
    
lst=[1,2,3]
print(id(lst))
modfy(lst)
print('list', lst)
```

### Kinds of arguments in python

### When you define a function, it is formal argument. While calling it is actual argument

### Actual argument: Position, Keyword, Default, Variable Length

```python
def person(name, age):
    print(name)
    print(age)
person('Manish','45')
person(age=34,name='Anand')
```

```python
def sum(a, *b):
    print(a)
    print(b)
    c=a
    for i in b:
        c=c+i
    print(c)
sum(2,3,4,5)
```

### Keyword variable length arguments(**kwargs)

```python
def person(name, *data):
    print(name)
    print(data)
    
person('Manish',26, 'Siraha','985467848')
```

```python
def person(name, **data):
    print(name)
    for i,j in data.items():
        print(i,j)
        #print(j)
    
person('Manish',age=26, hometown='Siraha',mob='985467848')
```

### Passing list to a function 

```python
def count(lst):
    even=0
    odd=0
    for i in lst:
        if i%2==0:
            even+=1
        else:
            odd+=1
    return even,odd

            
lst=[0,1,2,3,4,5,6]
even, odd=count(lst)
print("Even: {} and Odd: {}".format(even, odd))
```

### Fibonacci Sequence

```python
def fib(n):
    a=0
    b=1
    
    print(a)
    print(b)
        
    for i in range(n-2):
        c=a+b
        print(c)
        a=b
        b=c
fib(6)
    
```

```python
### Factorial
def factorial(n):
    if n==0:
        print(1)
    else:
        f=n
        c=n
        for i in range(n-1):
            f=f*(c-1)
            c=c-1
        print(f)
factorial(1)
            
            
        
```

### Recursion in python

#### Calling a function from itself

```python
import sys
sys.setrecursionlimit(900)
print(sys.getrecursionlimit())
c=0
def greet():
    global c
    c=c+1
    print('Hello',c)
    greet()
greet()
```

### Factorial using recursion

```python

def fact(n):
    if n==0:
        return 1
    else:
        return n*fact(n-1)

result=fact(5)
print(result)
```
